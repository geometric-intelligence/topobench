```python
import wandb

wandb_username = "gbg141-hopse"

models = [
    "topotune",
    "hopse_m",
    "gat",
    "gcn",
]

datasets = [
    "ZINC",
    "NCI109",
    "NCI1",
    "PROTEINS",
    "MUTAG",
    "cocitation_cora",
    "cocitation_citeseer",
    "cocitation_pubmed",
    "mantra_name",
    "mantra_orientation",
    "mantra_betti_numbers",
]

api = wandb.Api()
available_projects = {p.name for p in api.projects(wandb_username)}

experiments = []
for model in models:
    for dataset in datasets:
        name = f"{model}_{dataset}"
        if name in available_projects:
            experiments.append(name)
        else:
            print(f"⏭ Skipping {name} (not found in wandb)")

print(f"\n✓ Found {len(experiments)} experiments: {experiments}")

csv_filename = "hopse_hyperparam_runs.csv"

df = load_results_dataframe(
    wandb_username,
    experiments,
    csv_filename=csv_filename,
    force_load=False,
    save_csv=True,
    batch_size=500,  # Save to CSV every 500 new runs
    per_page=50,  # Smaller pages = more stable (less likely to timeout)
    fetch_recent_only=False,  # Fetch newest first, stop early (default: True)
    early_stop_threshold=100,  # Stop after 100 consecutive existing runs (default: 100)
)

print(f"Loaded {len(df)} runs")
```
